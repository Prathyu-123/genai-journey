```python
#groupby
# importing pandas library
import numpy as np

# Creating pandas dataframe
df = pd.DataFrame(
    [
        ("Corona Positive", 65, 99),
        ("Corona Negative", 52, 98.7),
        ("Corona Positive", 43, 100.1),
        ("Corona Positive", 26, 99.6),
        ("Corona Negative", 30, 98.1),
    ],
    
    index=["Patient 1", "Patient 2", "Patient 3",
           "Patient 4", "Patient 5"],
    
    columns=("Status", "Age(in Years)", "Temperature"),
)

# show dataframe
print(df)
```

```python
# Grouping with only status 
grouped1 = df.groupby("Status")

# Grouping with temperature and status
grouped3 = df.groupby(["Temperature", "Status"])
```

```python
print(grouped3.mean())

print(grouped1.sum())
```

```python
#merge()
import pandas as pd

# DataFrames to merge
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Age': [24, 27, 22]})

# Merge DataFrames on the 'ID' column using an inner join
merged_df = pd.merge(df1, df2, on='ID', how='left')
print(merged_df)
```

```python
#concat
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

vertical_concat = pd.concat([df1, df2], axis=0, ignore_index=True) # concatenating along rows

print("Vertical:")
display(vertical_concat)
```

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'C': [5, 6], 'D': [7, 8]})
horizontal_concat = pd.concat([df1, df2], axis=1)
display(horizontal_concat)
```

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [25, 30, 35, 40, 28],
    "Score": [85, 92, 78, 95, 88]
}
df = pd.DataFrame(data)

# Select row by index
print(df.loc[0])   # Row with index 0
print(df.iloc[2])  # Row at position 2

# Select multiple rows and columns
print(df.loc[0:2, ["Name", "Score"]])

# Boolean indexing (filter)
print(df[df["Score"] > 85])
```

```python
#Handling Missing Data
import numpy as np

# Introduce some NaN values
df.loc[1, "Score"] = np.nan
df.loc[3, "Age"] = np.nan
print(df)

# Check missing values
print(df.isnull())

# Drop rows with missing values
print(df.dropna())

# Fill missing values
print(df.fillna({"Score": 0, "Age": df["Age"].mean()}))
```

```python
#Sorting
# Sort by Score
print(df.sort_values(by="Score", ascending=False))

# Sort by index
print(df.sort_index())
```

```python
#Basic Statistics
print("Mean Age:", df["Age"].mean())
print("Median Score:", df["Score"].median())
print("Max Score:", df["Score"].max())
print("Min Age:", df["Age"].min())

# Correlation between numeric columns
print(df.corr())
```
